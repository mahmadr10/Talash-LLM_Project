# TALASH Milestone 2 Continuation

This notebook continues the milestone 1 CV extraction work and demonstrates the milestone 2 analysis layer on top of the existing structured sample data.

```python
import json
import sys
from pathlib import Path

project_root = Path.cwd().resolve()
milestone_2_dir = project_root / 'milestone_2'
if str(milestone_2_dir) not in sys.path:
    sys.path.insert(0, str(milestone_2_dir))

from milestone2 import Milestone2Analysis

sample_path = milestone_2_dir / 'sample_cv_data.json'
payload = json.loads(sample_path.read_text(encoding='utf-8'))
candidate = payload['candidates'][0]
analysis = Milestone2Analysis(candidate).run_all_analyses()
analysis.keys()
```

```python
summary_rows = []
for candidate in payload['candidates']:
    result = Milestone2Analysis(candidate).run_all_analyses()
    summary_rows.append({
        'candidate': candidate['candidates']['full_name'],
        'highest_qualification': result.get('education_analysis', {}).get('highest_qualification', 'N/A') if isinstance(result.get('education_analysis'), dict) else 'N/A',
        'experience_roles': result.get('experience_analysis', {}).get('employment_history_count', 0) if isinstance(result.get('experience_analysis'), dict) else 0,
        'missing_items': result.get('candidate_summary', {}).get('missing_items', 0) if isinstance(result.get('candidate_summary'), dict) else 0,
        'alignment_ratio': result.get('skill_alignment', {}).get('alignment_ratio', '0%') if isinstance(result.get('skill_alignment'), dict) else '0%'
    })
summary_rows
```
